```python
# finds all the div, the h6 which there names, and p's which are there titles that are of the specified class, this one is the class that 
# holds the professors names, i limited it 50 since two professors are retired from the list
test = soup.find_all(['div', 'h6', 'p', 'strong'], attrs={"class":['small-12 medium-6 large-3 columns faculty-wrapper']}, limit=50)
```

```python
# this was a test 
soup.find_all('div', class_='small-12 medium-6 large-3 columns faculty-wrapper')
```

```python
prof_name = []
prof_title = []

# add the profrssors name and title to lists
for person in test:
    prof = person.h6.text
    #print(prof)
    prof_name.append(prof)
    title = person.p.text 
    #print(t)
    prof_title.append(title)

#makes the lists to a data frame
df = pd.DataFrame({
    'Professor': prof_name,
    'Title': prof_title
})

```

```python

# Split the 'Professor' column into 'First Name' and 'Last Name'
df[['First Name', 'Last Name']] = df['Professor'].str.split(' ', n=1, expand=True)

# drop's the original 'Professor' 
df = df.drop(columns=['Professor'])
```

```python
df
```

```python
# changes the order of the columns
df = df[['Last Name', 'First Name', 'Title']]
```

```python
# splits everything after the ',' in the last name column
df['Last Name'] = df['Last Name'].str.split(',', n=1).str[0]
df
```

```python
# makes a file with the list of professors
name = 'new_professors.csv'

df.to_csv(name, index=False)
```
